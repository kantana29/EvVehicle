```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ev = pd.read_csv('Ev_state.csv')
ev.head()
```

```python
ev.isnull().sum()
```

```python
#fill ,missing values

ev["Total Electric Vehicle"].fillna(ev["Total Electric Vehicle"].mode()[0],inplace=True)
ev["Total Non-Electric Vehicle"].fillna(ev["Total Non-Electric Vehicle"].mode()[0],inplace=True)
ev["Total"].fillna(ev["Total"].mode()[0],inplace=True)
```

```python
ev.isnull().sum()
```

```python
ev["Total Electric Vehicle"]=pd.to_numeric(ev["Total Electric Vehicle"],errors='coerce')
```

```python
ev.info()
```

```python
X=ev["State Name"]
Y=ev["Total Electric Vehicle"]
```

# A bar graph is a graphical representation of information. It uses bars that extend to different heights to depict value. Bar graphs can be created with vertical bars, horizontal bars

```python
plt.barh(X,Y,color='b',align='center')
plt.title("State Wise Electric Vehicle Shows")
plt.xlabel("Number of Ev's in India")
plt.ylabel("States")
plt.figure(figsize=(40,10))
plt.xticks(rotation=70)  # Rotate the y-tick labels by 45 degrees
plt.show()
```
