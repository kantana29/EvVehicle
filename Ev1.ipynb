```python
# Count the frequency of each category
style_counts = ev['Style'].value_counts()
# Create a list of colors for each category
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

# Create a bar plot
plt.bar(style_counts.index, style_counts.values, color=colors)

# Set the labels and title
plt.xlabel('Style')
plt.ylabel('Count')
plt.title('Different style of Car Capacity')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Count the frequency of each category
car_counts = ev['Car'].value_counts()
# Create a list of colors for each category
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

# Create a bar plot
plt.bar(car_counts.index, car_counts.values, color=colors)

# Set the labels and title
plt.xlabel('Car model')
plt.ylabel('Count')
plt.title('Different brands of Car Capacity')
plt.xticks(rotation=70)

# Display the plot
plt.show()
```

```python
# Extract the numeric value from the 'Range' column
ev['Range'] = ev['Range'].str.extract('(\d+)', expand=False).astype(float)
```

```python
ev.head()
```

```python
# Calculate the frequency/count of each range
range_counts = ev['Range'].value_counts()
# Create a pie plot
plt.pie(range_counts.values, labels=range_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of Range')

# Display the plot
plt.show()
```

```python
# Sort the DataFrame by 'Range' column in descending order and select the top 10 cars
top_10_cars = ev.nlargest(10, 'Range')

# Create a list of colors for each category
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

# Create a bar plot for the top 10 cars and their ranges
plt.bar(top_10_cars['Car'], top_10_cars['Range'], color=colors)

# Set the labels and title
plt.xlabel('Car')
plt.ylabel('Range (Km/Full Charge)')
plt.title('Top 10 Cars by Range')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=70)

# Display the plot
plt.show()
```
