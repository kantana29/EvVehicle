```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ev = pd.read_csv('ev_india.csv')
ev.head()
```

```python
# Group the data by brand and calculate the maximum top speed
brand_top_speeds = ev.groupby('Brand')['TopSpeed_KmH'].max().nlargest(10)

# Create a list of colors for each category
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

# Create a bar plot for the top speeds by brand
plt.bar(brand_top_speeds.index, brand_top_speeds.values, color=colors)

# Set the labels and title
plt.xlabel('Brand')
plt.ylabel('Top Speed (Km/H)')
plt.title('Top 10 Car Brands by Top Speed')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Count the frequency of each RapidCharge category
rapid_charge_counts = ev['RapidCharge'].value_counts()

# Create a pie plot
plt.pie(rapid_charge_counts.values, labels=rapid_charge_counts.index, autopct='%1.1f%%')

# Set the title
plt.title('Distribution of Rapid Charge')

# Display the plot
plt.show()
```

```python
# Count the frequency of each body style category
body_style_counts = ev['BodyStyle'].value_counts()

# Create a list of colors for each category
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

# Create a bar plot
plt.bar(body_style_counts.index, body_style_counts.values, color=colors)

# Set the labels and title
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.title('Distribution of Body Style')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = ev.corr()

# Create a heatmap using the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Set the title
plt.title('Correlation Heatmap')

# Display the plot
plt.show()
```
